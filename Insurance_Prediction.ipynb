```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pickle
```

```python
df= pd.read_csv("insurance.csv")
print(df.isnull().sum())
print(df['sex'].value_counts())
print(df['smoker'].value_counts())
print(df['region'].value_counts())



x=df.iloc[:,:-1].values
y=df.iloc[:,-1].values

print(x[:10])
print(y[:10])
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
columnss=[1,4,5]
ct1= ColumnTransformer(transformers=[('encoder',OneHotEncoder(),columnss)],remainder='passthrough')
x=ct1.fit_transform(x)
print(x[:10])
```

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest= train_test_split(x,y,test_size=0.2,random_state=0)


print(xtrain[:10])
print(xtest[:10])
print(ytrain[:10])
print(ytest[:10])
```

```python
from sklearn.preprocessing import StandardScaler
st= StandardScaler()
xtrain[:,[-2,-3]]=st.fit_transform(xtrain[:,[-2,-3]])
xtest[:,[-2,-3]]=st.transform(xtest[:,[-2,-3]])

print(xtrain[:10])
print(xtest[:10])
```

```python
st1=StandardScaler()
ytrain=st1.fit_transform(ytrain.reshape(-1,1))
ytest=st1.transform(ytest.reshape(-1,1))
```

```python
with open('column_transformer.pkl', 'wb') as f:
    pickle.dump(ct1, f)

with open('feature_scaler.pkl', 'wb') as f:
    pickle.dump(st, f)

with open('target_scaler.pkl', 'wb') as f:
    pickle.dump(st1, f)
```

```python
from sklearn.svm import SVR
regressor= SVR(kernel='rbf')
regressor.fit(xtrain,ytrain)
```

```python
y_pred= regressor.predict(xtest)
print(st1.inverse_transform(y_pred.reshape(-1,1)))
```

```python
from sklearn.metrics import r2_score
print(r2_score(ytest,y_pred))

```
